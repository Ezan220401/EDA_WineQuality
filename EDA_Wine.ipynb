```python
import numpy as np

def drop_outlier(x,y):
  df=pd.DataFrame(x, columns=['x'])
  df['y']=y

  q1=np.percentile(df['x'], 25, interpolation='midpoint')
  q3=np.percentile(df['x'], 75, interpolation='midpoint')

  IQR=q3-q1

  # Menentukan upper dan lower limit
  upper=np.where(df['x'] >= (q3 + 1.5 * IQR))
  lower=np.where(df['x'] >= (q3 + 1.5 * IQR))

  # Drop value
  df.drop(upper[0], inplace=True)
  df.drop(lower[0], inplace=True)
  df.reset_index(drop=True, inplace=True)

  return df['x'].values, df['y'].values
```
